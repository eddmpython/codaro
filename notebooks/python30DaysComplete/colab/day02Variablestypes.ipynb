## 왜 중요한가

프로그램은 값을 저장하고 다시 꺼내 쓰는 일의 반복입니다. 변수는 그 반복을 시작하는 첫 도구입니다.

## 생각 모델

변수는 값에 붙이는 이름표입니다. 이름표를 바꾸면 다음 줄에서 꺼내 쓰는 값도 달라집니다.
