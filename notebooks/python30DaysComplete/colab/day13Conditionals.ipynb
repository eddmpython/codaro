## 오늘의 목표

if, elif, else로 조건에 따라 다른 코드를 실행합니다.

## 왜 중요한가

프로그램이 상황에 맞게 다르게 행동하려면 조건문이 필요합니다.

## 생각 모델

조건문은 갈림길입니다. 위에서부터 조건을 검사하고 처음 맞는 길 하나를 실행합니다.

## 오늘 배우는 것

- if로 조건 분기
- elif와 else로 다중 분기
- 중첩 조건문과 삼항연산자
- 실전 조건 처리

## 학습 방법

1. 설명을 먼저 읽습니다.
2. 바로 아래 코드 셀을 실행합니다.
3. 출력이 설명과 어떻게 연결되는지 한 문장으로 말합니다.
4. 연습 셀에는 예제를 보지 않고 직접 다시 작성합니다.

## 오늘의 범위

- 오늘 새로 배우는 개념: if, elif, else, 한 줄 조건식, 중첩 if
- 이미 써도 되는 개념: 지금까지 배운 자료형, 지금까지 배운 연산자
- 오늘은 일부러 쓰지 않는 개념: 직접 만드는 함수, 반복문, import

범위를 좁히는 이유는 간단합니다. 처음 배우는 사람은 한 번에 많은 문법을 보면 어디서 막혔는지 찾기 어렵습니다.

## 오늘의 학습 전략

- 코딩 전: 각 조건이 참인지 거짓인지 손으로 먼저 판단합니다.
- 실행 중: 값을 바꾸며 어떤 가지가 실행되는지 출력으로 확인합니다.
- 연습 초점: 점수, 나이, 상태값을 기준으로 다른 메시지를 출력합니다.
- 막히면: 원하지 않는 가지가 실행되면 조건의 순서와 들여쓰기를 확인합니다.
