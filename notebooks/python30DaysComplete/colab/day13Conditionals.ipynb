## if 기본

*조건이 참일 때 실행*

if문은 조건이 True일 때만 코드를 실행합니다. if 조건: 형식으로 쓰며, 조건 뒤에 콜론(:)을 붙이고 실행할 코드는 들여쓰기합니다. 조건이 False면 if 블록을 건너뜁니다.

- if 조건: 형식
- 조건이 True면 실행
- 조건이 False면 건너뜀
- 들여쓰기로 블록 구분

### if 기본

조건이 참이면 실행됩니다.

```python
age = 20
if age >= 18:
    result = 'adult'
result
```

### if 조건 거짓

조건이 거짓이면 건너뜁니다.

```python
score = 50
if score >= 90:
    grade = 'A'
score
```

### if 변수 수정

조건에 따라 변수를 수정합니다.

```python
cost = 10000
if cost > 5000:
    cost = cost - 1000
cost
```

> **팁**
>
> if문의 조건은 비교 연산자, 논리 연산자, in 연산자 등을 사용할 수 있습니다.

## if-else

*조건에 따른 양방향 분기*

else는 if 조건이 False일 때 실행됩니다. if-else를 사용하면 조건에 따라 둘 중 하나가 반드시 실행됩니다. else 뒤에도 콜론(:)을 붙이고 들여쓰기합니다.

- if-else로 양방향 분기
- 조건이 True면 if 블록
- 조건이 False면 else 블록
- 둘 중 하나는 반드시 실행

### if-else 기본

조건에 따라 다른 값을 할당합니다.

```python
num = 7
if num % 2 == 0:
    parity = 'even'
else:
    parity = 'odd'
parity
```

### 성인 판별

나이에 따라 성인 여부를 판별합니다.

```python
years = 16
if years >= 18:
    status = 'adult'
else:
    status = 'minor'
status
```

### 할인 적용

가격에 따라 할인을 적용합니다.

```python
amount = 15000
if amount >= 10000:
    final = amount - 2000
else:
    final = amount
final
```

> **팁**
>
> if-else는 둘 중 하나를 반드시 선택해야 할 때 유용합니다.

## if-elif-else

*다중 조건 분기*

elif는 else if의 줄임말로, 여러 조건을 순차적으로 검사합니다. if-elif-else를 사용하면 3개 이상의 경우를 나눌 수 있습니다. 위에서부터 조건을 검사하여 처음 True인 블록만 실행됩니다.

- elif로 다중 조건 검사
- 위에서부터 순서대로 검사
- 첫 번째 True 블록만 실행
- 모두 False면 else 실행

### 성적 등급

점수에 따라 등급을 부여합니다.

```python
point = 85
if point >= 90:
    level = 'A'
elif point >= 80:
    level = 'B'
elif point >= 70:
    level = 'C'
else:
    level = 'F'
level
```

### 온도 분류

온도에 따라 분류합니다.

```python
temp = 25
if temp >= 30:
    weather = 'hot'
elif temp >= 20:
    weather = 'warm'
elif temp >= 10:
    weather = 'cool'
else:
    weather = 'cold'
weather
```

### 요금 계산

거리에 따라 요금을 계산합니다.

```python
distance = 15
if distance <= 5:
    fare = 1000
elif distance <= 10:
    fare = 2000
elif distance <= 20:
    fare = 3000
else:
    fare = 5000
fare
```

> **팁**
>
> elif는 여러 개 사용할 수 있으며, else는 생략 가능합니다.

## 비교 연산자와 조건문

*크기 비교로 조건 판별*

비교 연산자(==, !=, >, <, >=, <=)는 조건문에서 가장 많이 사용됩니다. 두 값을 비교하여 True/False를 반환하며, 이를 if문의 조건으로 사용합니다.

- ==: 같음
- !=: 다름
- >, <: 크기 비교
- >=, <=: 이상, 이하

### 같음 비교

값이 같은지 비교합니다.

```python
password = '1234'
if password == '1234':
    access = 'granted'
else:
    access = 'denied'
access
```

### 다름 비교

값이 다른지 비교합니다.

```python
stock = 0
if stock != 0:
    available = True
else:
    available = False
available
```

### 크기 비교

크기를 비교합니다.

```python
speed = 120
if speed > 100:
    warning = 'overspeed'
else:
    warning = 'normal'
warning
```

> **팁**
>
> 문자열도 비교 연산자로 비교할 수 있습니다(사전 순서).

## 논리 연산자 and

*여러 조건을 모두 만족*

and 연산자는 양쪽 조건이 모두 True일 때만 True를 반환합니다. 여러 조건을 동시에 만족해야 할 때 사용합니다. 조건1 and 조건2 형식입니다.

- 양쪽 모두 True일 때만 True
- 하나라도 False면 False
- 여러 조건 동시 검사
- 조건1 and 조건2 형식

### and 기본

두 조건을 모두 만족하는지 검사합니다.

```python
userAge = 25
isMember = True
if userAge >= 18 and isMember:
    rate = 0.2
else:
    rate = 0
rate
```

### 범위 검사

값이 특정 범위 안에 있는지 검사합니다.

```python
num = 55
if num >= 50 and num <= 100:
    inRange = True
else:
    inRange = False
inRange
```

### 자격 검사

여러 조건을 모두 만족하는지 검사합니다.

```python
experience = 3
degree = True
if experience >= 2 and degree:
    qualified = 'yes'
else:
    qualified = 'no'
qualified
```

> **팁**
>
> 범위 검사는 50 <= value <= 100 형식으로도 가능합니다.

## 논리 연산자 or

*조건 중 하나만 만족*

or 연산자는 양쪽 조건 중 하나라도 True면 True를 반환합니다. 여러 조건 중 하나만 만족해도 될 때 사용합니다. 조건1 or 조건2 형식입니다.

- 하나라도 True면 True
- 모두 False일 때만 False
- 대안 조건 검사
- 조건1 or 조건2 형식

### or 기본

두 조건 중 하나만 만족해도 됩니다.

```python
weekend = False
holiday = True
if weekend or holiday:
    day = 'off'
else:
    day = 'work'
day
```

### 자격 대안

여러 자격 중 하나만 있어도 됩니다.

```python
card = False
cash = True
if card or cash:
    pay = True
else:
    pay = False
pay
```

### 경고 조건

여러 조건 중 하나라도 위험하면 경고합니다.

```python
battery = 15
signal = 80
if battery < 20 or signal < 30:
    alert = 'warning'
else:
    alert = 'normal'
alert
```

> **팁**
>
> or는 첫 번째 True를 만나면 나머지 조건은 검사하지 않습니다.

## 논리 연산자 not

*조건 반대로 뒤집기*

not 연산자는 조건의 결과를 반대로 뒤집습니다. True는 False로, False는 True로 바꿉니다. not 조건 형식으로 사용합니다.

- 조건 결과 반대로
- True → False
- False → True
- not 조건 형식

### not 기본

조건의 반대를 검사합니다.

```python
logged = False
if not logged:
    action = 'login required'
else:
    action = 'proceed'
action
```

### not과 비교

not과 비교 연산자를 함께 사용합니다.

```python
count = 5
if not count == 0:
    ready = True
else:
    ready = False
ready
```

### not과 논리 연산

not과 and/or를 함께 사용합니다.

```python
premium = False
trial = False
if not (premium or trial):
    plan = 'free'
else:
    plan = 'paid'
plan
```

> **팁**
>
> not count == 0 대신 count != 0을 사용하는 것이 더 읽기 쉽습니다.

## in 연산자와 조건문

*포함 여부 검사*

in 연산자는 값이 리스트, 튜플, 집합, 딕셔너리, 문자열에 포함되어 있는지 검사합니다. 값 in 컬렉션 형식으로 사용하며, True/False를 반환합니다.

- 값 in 컬렉션 형식
- 포함되면 True
- 없으면 False
- 리스트, 튜플, 집합, 문자열 등

### 리스트 포함 검사

값이 리스트에 포함되어 있는지 검사합니다.

```python
users = ['admin', 'user1', 'user2']
user = 'user1'
if user in users:
    permission = 'granted'
else:
    permission = 'denied'
permission
```

### 문자열 포함 검사

문자가 문자열에 포함되어 있는지 검사합니다.

```python
email = 'test@example.com'
if '@' in email:
    ok = True
else:
    ok = False
ok
```

### 딕셔너리 키 검사

키가 딕셔너리에 있는지 검사합니다.

```python
config = {'host': 'localhost', 'port': 8080}
if 'timeout' in config:
    timeout = config['timeout']
else:
    timeout = 30
timeout
```

> **팁**
>
> not in 연산자로 포함되지 않은 것을 검사할 수 있습니다.

## 중첩 조건문

*조건 안에 조건*

if문 안에 다시 if문을 넣을 수 있습니다. 이를 중첩 조건문이라고 하며, 복잡한 조건을 단계적으로 검사할 때 사용합니다. 들여쓰기 단계가 늘어납니다.

- if 안에 if 중첩
- 단계적 조건 검사
- 들여쓰기 단계 증가
- 복잡한 조건 처리

### 중첩 if 기본

조건을 단계적으로 검사합니다.

```python
driverAge = 25
hasLicense = True
if driverAge >= 18:
    if hasLicense:
        canDrive = 'yes'
    else:
        canDrive = 'no license'
else:
    canDrive = 'too young'
canDrive
```

### 3단계 검사

여러 단계로 조건을 검사합니다.

```python
wallet = 50000
cost = 30000
inStock = True
if wallet >= cost:
    if inStock:
        outcome = 'purchase'
    else:
        outcome = 'out of stock'
else:
    outcome = 'insufficient'
outcome
```

### 등급과 보너스

등급에 따라 보너스를 차등 지급합니다.

```python
grade = 'gold'
purchase = 100000
if grade == 'gold':
    if purchase >= 50000:
        bonus = 10000
    else:
        bonus = 5000
else:
    bonus = 0
bonus
```

> **팁**
>
> 중첩이 깊어지면 and 연산자로 단순화할 수 있습니다.

## 삼항 연산자

*한 줄로 조건 분기*

삼항 연산자는 if-else를 한 줄로 표현합니다. 값1 if 조건 else 값2 형식으로 쓰며, 조건이 True면 값1, False면 값2를 반환합니다. 간단한 조건 분기에 유용합니다.

- 값1 if 조건 else 값2 형식
- 조건이 True면 값1
- 조건이 False면 값2
- 한 줄로 간결하게

### 삼항 연산자 기본

간단한 조건을 한 줄로 표현합니다.

```python
n = 10
category = 'even' if n % 2 == 0 else 'odd'
category
```

### 최소값 적용

조건에 따라 값을 선택합니다.

```python
qty = 3
limit = 5
final = qty if qty >= limit else limit
final
```

### 할인율 결정

금액에 따라 할인율을 결정합니다.

```python
total = 120000
rate = 0.1 if total >= 100000 else 0.05
discounted = total * (1 - rate)
discounted
```

> **팁**
>
> 삼항 연산자는 간단한 조건에만 사용하세요. 복잡하면 일반 if-else가 더 읽기 쉽습니다.
