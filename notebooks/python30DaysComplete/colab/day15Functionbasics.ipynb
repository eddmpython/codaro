## 함수 정의

*def로 함수 만들기*

함수는 특정 작업을 수행하는 코드 블록입니다. def 함수명(): 형식으로 정의하며, 들여쓰기로 함수 본문을 작성합니다. 함수를 정의한 후 함수명()으로 호출하여 실행합니다.

- def 함수명(): 형식
- 들여쓰기로 본문 작성
- 함수명()으로 호출
- 코드 재사용 가능

### 함수 정의

간단한 함수를 정의합니다.

```python
def greet():
    msg = 'Hello'
    return msg

greet()
```

### 함수 호출

함수를 여러 번 호출할 수 있습니다.

```python
def getMessage():
    return 'Welcome'

first = getMessage()
second = getMessage()
first, second
```

### 함수 결과 사용

함수 결과를 변수에 저장하고 사용합니다.

```python
def getNumber():
    return 42

num = getNumber()
twice = num * 2
twice
```

## return 문

*값 반환하기*

return은 함수의 결과값을 반환합니다. return 값 형식으로 사용하며, 함수 실행을 즉시 종료하고 값을 호출한 곳으로 돌려줍니다. return이 없으면 None을 반환합니다.

- return 값 형식
- 함수 결과 반환
- 함수 즉시 종료
- return 없으면 None

### return 기본

계산 결과를 반환합니다.

```python
def add():
    answer = 10 + 20
    return answer

add()
```

### return 즉시

계산식을 바로 반환합니다.

```python
def multiply():
    return 5 * 6

multiply()
```

### 조건부 return

조건에 따라 다른 값을 반환합니다.

```python
def checkValue():
    value = 15
    if value > 10:
        return 'big'
    else:
        return 'small'

checkValue()
```

> **팁**
>
> return을 만나면 함수가 즉시 종료되므로 위치에 주의하세요.

## 매개변수 1개

*값을 받는 함수*

매개변수는 함수가 받는 입력값입니다. def 함수명(매개변수): 형식으로 정의하며, 호출시 전달하는 값을 인자라고 합니다. 매개변수를 사용하면 함수를 더 유연하게 만들 수 있습니다.

- def 함수명(매개변수): 형식
- 매개변수: 함수가 받는 변수
- 인자: 호출시 전달하는 값
- 함수 유연성 증가

### 매개변수 기본

매개변수를 받아 처리합니다.

```python
def double(num):
    return num * 2

double(5)
```

### 문자열 매개변수

문자열을 받아 처리합니다.

```python
def sayHi(name):
    return 'Hello ' + name

sayHi('Alice')
```

### 조건과 매개변수

매개변수를 조건문에서 사용합니다.

```python
def isAdult(age):
    if age >= 18:
        return True
    else:
        return False

isAdult(25)
```

>
> 매개변수명은 함수 안에서만 유효한 지역 변수입니다.

## 매개변수 여러 개

*여러 값 받기*

함수는 여러 개의 매개변수를 받을 수 있습니다. 쉼표로 구분하여 나열하며, 호출시에도 같은 순서로 인자를 전달해야 합니다.

- def 함수명(매개변수1, 매개변수2): 형식
- 쉼표로 구분
- 순서대로 전달
- 여러 값 조합 가능

### 두 개 매개변수

두 개의 매개변수를 받아 계산합니다.

```python
def addTwo(a, b):
    return a + b

addTwo(10, 20)
```

### 세 개 매개변수

세 개의 매개변수를 받아 계산합니다.

```python
def average(x, y, z):
    return (x + y + z) / 3

average(80, 90, 85)
```

### 문자열 조합

여러 문자열을 조합합니다.

```python
def fullName(first, last):
    return first + ' ' + last

fullName('John', 'Doe')
```

## 여러 값 반환

*튜플로 반환*

함수는 여러 값을 반환할 수 있습니다. 쉼표로 구분하여 반환하면 튜플로 묶여서 반환됩니다. 호출한 곳에서 여러 변수로 받을 수 있습니다.

- return 값1, 값2 형식
- 튜플로 반환됨
- 여러 변수로 받기 가능
- 관련된 값 함께 반환

### 두 값 반환

두 개의 값을 반환합니다.

```python
def minMax(a, b):
    if a < b:
        return a, b
    else:
        return b, a

minMax(10, 5)
```

### 계산 결과들

여러 계산 결과를 반환합니다.

```python
def calculate(num):
    double = num * 2
    triple = num * 3
    return double, triple

calculate(5)
```

### 분리해서 받기

반환된 값을 각각 변수에 저장합니다.

```python
def getInfo():
    name = 'Python'
    version = 3.9
    return name, version

lang, ver = getInfo()
lang
```

>
> 여러 값을 반환하면 자동으로 튜플이 됩니다.

## return 없는 함수

*작업만 수행*

함수는 return이 없어도 됩니다. return이 없으면 None을 반환합니다. 주로 작업만 수행하고 결과가 필요없을 때 사용합니다.

- return 생략 가능
- None 자동 반환
- 작업 수행용
- 부수 효과만 목적

### return 없음

return이 없으면 None을 반환합니다.

```python
def doSomething():
    x = 10 + 20

output = doSomething()
output
```

### 조건부 return

조건에 따라 return이 없을 수도 있습니다.

```python
def checkPositive(num):
    if num > 0:
        return 'positive'

checkPositive(-5)
```

명시적으로 None을 반환할 수도 있습니다.

```python
def getValue(flag):
    if flag:
        return 100
    return None

getValue(False)
```

## 함수와 조건문

*조건에 따른 처리*

함수 안에서 조건문을 사용하여 매개변수에 따라 다른 결과를 반환할 수 있습니다. 함수의 유연성이 크게 증가합니다.

- 함수 내부에 조건문
- 매개변수 기반 분기
- 다양한 경우 처리
- 유연한 함수 구현

### 등급 판정

점수에 따라 등급을 반환합니다.

```python
def getGrade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'F'

getGrade(85)
```

### 절대값

숫자의 절대값을 반환합니다.

```python
def absolute(num):
    if num >= 0:
        return num
    else:
        return -num

absolute(-15)
```

### 할인 계산

금액에 따라 할인을 적용합니다.

```python
def applyDiscount(price):
    if price >= 100000:
        return price * 0.9
    elif price >= 50000:
        return price * 0.95
    else:
        return price

applyDiscount(80000)
```

> **팁**
>
> 조건문으로 다양한 입력에 대응할 수 있습니다.

## 함수와 반복문

*반복 처리 함수*

함수 안에서 반복문을 사용하여 리스트나 문자열을 처리할 수 있습니다. 반복 로직을 함수로 만들면 재사용이 쉬워집니다.

- 함수 내부에 반복문
- 리스트/문자열 처리
- 반복 로직 재사용
- 코드 간결화

### 리스트 합계

리스트의 모든 요소를 더합니다.

```python
def sumList(numbers):
    acc = 0
    for num in numbers:
        acc = acc + num
    return acc

sumList([1, 2, 3, 4, 5])
```

### 문자 개수

문자열에서 특정 문자의 개수를 셉니다.

```python
def countChar(text, target):
    cnt = 0
    for char in text:
        if char == target:
            cnt = cnt + 1
    return cnt

countChar('hello world', 'o')
```

### 최대값 찾기

리스트에서 최대값을 찾습니다.

```python
def findMax(values):
    biggest = values[0]
    for item in values:
        if item > biggest:
            biggest = item
    return biggest

findMax([15, 42, 8, 23, 56])
```

> **팁**
>
> 복잡한 반복 로직을 함수로 만들면 코드가 깔끔해집니다.

## 함수 재사용

*같은 함수 여러 번 사용*

함수의 가장 큰 장점은 재사용입니다. 한 번 정의한 함수를 여러 곳에서 다른 인자로 호출할 수 있습니다. 코드 중복을 줄이고 유지보수가 쉬워집니다.

- 한 번 정의, 여러 번 사용
- 다른 인자로 호출
- 코드 중복 감소
- 유지보수 용이

### 여러 번 호출

같은 함수를 다른 인자로 호출합니다.

```python
def square(n):
    return n * n

sq3 = square(3)
sq5 = square(5)
sq7 = square(7)
sq3, sq5, sq7
```

### 함수 조합

함수 결과를 다시 함수에 전달합니다.

```python
def addTen(num):
    return num + 10

chained = addTen(5)
chained = addTen(chained)
chained
```

### 반복문에서 호출

반복문 안에서 함수를 호출합니다.

```python
def triple(x):
    return x * 3

results = []
for i in range(1, 5):
    results.append(triple(i))
results
```

> **팁**
>
> 자주 사용하는 로직은 함수로 만들어 재사용하세요.
