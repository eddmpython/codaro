## try-except 기본

*에러 처리하기*

try-except 문으로 에러를 처리합니다. try 블록에서 에러가 발생하면 except 블록이 실행됩니다. 프로그램이 중단되지 않고 계속 실행됩니다.

- try: 실행할 코드
- except: 에러 처리
- 에러 발생시 except 실행
- 프로그램 중단 방지

### 기본 예외 처리

에러가 발생해도 프로그램이 계속 실행됩니다.

```python
try:
    result = 10 / 0
except:
    result = 'Error occurred'

result
```

### 정상 실행

에러가 없으면 except는 실행되지 않습니다.

```python
try:
    value = 10 / 2
except:
    value = 'Error'

value
```

### 함수에서 사용

함수 안에서도 예외를 처리할 수 있습니다.

```python
def safeDivide(a, b):
    try:
        return a / b
    except:
        return 0

safeDivide(10, 0)
```

> **팁**
>
> try-except는 예상되는 에러를 처리할 때 사용합니다.

## 특정 예외 처리

*예외 타입 지정*

except 뒤에 예외 타입을 지정할 수 있습니다. 특정 에러만 처리하고 다른 에러는 그대로 발생시킵니다. 더 정확한 에러 처리가 가능합니다.

- except 예외타입:
- 특정 에러만 처리
- 다른 에러는 발생
- 정확한 처리 가능

### ZeroDivisionError

0으로 나누는 에러만 처리합니다.

```python
try:
    num = 10 / 0
except ZeroDivisionError:
    num = 'Cannot divide by zero'

num
```

### ValueError

잘못된 값 변환 에러를 처리합니다.

```python
try:
    converted = int('abc')
except ValueError:
    converted = 0

converted
```

### IndexError

인덱스 범위 에러를 처리합니다.

```python
try:
    items = [1, 2, 3]
    item = items[10]
except IndexError:
    item = 'Index out of range'

item
```

> **팁**
>
> 예외 타입을 지정하면 의도하지 않은 에러를 놓치지 않습니다.

## 여러 예외 처리

*다양한 에러 대응*

여러 except 블록으로 다양한 예외를 처리할 수 있습니다. 위에서부터 순서대로 검사하며, 처음 일치하는 except가 실행됩니다.

- 여러 except 블록
- 위에서부터 검사
- 첫 일치만 실행
- 다양한 에러 처리

### 여러 예외

여러 종류의 에러를 각각 처리합니다.

```python
try:
    val = int('10')
    ans = val / 2
except ValueError:
    ans = 'Invalid number'
except ZeroDivisionError:
    ans = 'Cannot divide'

ans
```

### 튜플로 묶기

여러 예외를 하나의 except로 처리합니다.

```python
try:
    data = int('5')
    output = data / 1
except (ValueError, ZeroDivisionError):
    output = 'Error'

output
```

### 순서 중요

위에서부터 검사하므로 순서가 중요합니다.

```python
try:
    nums = [1, 2, 3]
    elem = nums[1]
except IndexError:
    elem = 'Index error'
except:
    elem = 'Other error'

elem
```

> **팁**
>
> 구체적인 예외를 먼저, 일반적인 예외를 나중에 배치하세요.

## finally 절

*항상 실행되는 코드*

finally 블록은 예외 발생 여부와 관계없이 항상 실행됩니다. 파일 닫기, 리소스 해제 등 정리 작업에 사용합니다.

- 항상 실행됨
- 에러 상관없이
- 정리 작업용
- try-except-finally

### finally 기본

finally는 항상 실행됩니다.

```python
try:
    calc = 10 / 2
except:
    calc = 0
finally:
    status = 'Completed'

calc, status
```

### 에러시에도 실행

에러가 발생해도 finally는 실행됩니다.

```python
try:
    res = 10 / 0
except ZeroDivisionError:
    res = 'Error'
finally:
    msg = 'Done'

res, msg
```

### 리소스 정리

파일 등의 리소스를 정리합니다.

```python
opened = False
try:
    f = open('test.txt', 'w')
    opened = True
    f.write('Data')
except:
    content = 'Error'
finally:
    if opened:
        f.close()
    content = 'Closed'

content
```

> **팁**
>
> with 문을 사용하면 finally 없이도 자동으로 정리됩니다.

## raise로 예외 발생

*의도적으로 에러 발생*

raise 키워드로 예외를 발생시킬 수 있습니다. 조건이 맞지 않을 때 강제로 에러를 내어 프로그램을 중단하거나 예외를 전달합니다.

- raise 예외타입(메시지)
- 의도적 에러 발생
- 조건 검증
- 예외 전달

### ValueError 발생

잘못된 값일 때 예외를 발생시킵니다.

```python
def checkAge(age):
    if age < 0:
        raise ValueError('Age cannot be negative')
    return age

try:
    valid = checkAge(-5)
except ValueError:
    valid = 'Invalid age'

valid
```

### 조건 검증

조건을 만족하지 않으면 예외를 발생시킵니다.

```python
def divide(a, b):
    if b == 0:
        raise ZeroDivisionError('Divisor cannot be zero')
    return a / b

try:
    quotient = divide(10, 0)
except ZeroDivisionError:
    quotient = 'Error'

quotient
```

### 범위 검사

범위를 벗어나면 예외를 발생시킵니다.

```python
def setScore(score):
    if score < 0 or score > 100:
        raise ValueError('Score out of range')
    return score

try:
    final = setScore(150)
except ValueError:
    final = 0

final
```

> **팁**
>
> raise는 함수의 입력값 검증에 유용합니다.

## 예외 정보

*에러 메시지 받기*

except 예외타입 as 변수: 형식으로 예외 객체를 받을 수 있습니다. 예외 객체는 에러 메시지와 정보를 담고 있습니다.

- except 타입 as 변수:
- 예외 객체 받기
- 에러 메시지 확인
- 디버깅에 유용

### 예외 객체

예외 객체에서 메시지를 가져옵니다.

```python
try:
    bad = 10 / 0
except ZeroDivisionError as err:
    bad = str(err)

bad
```

### 변환 에러

ValueError의 메시지를 확인합니다.

```python
try:
    num = int('invalid')
except ValueError as error:
    num = str(error)

num
```

### 함수에서 사용

함수에서 예외 정보를 반환합니다.

```python
def safeConvert(text):
    try:
        return int(text)
    except ValueError as e:
        return str(e)

safeConvert('abc')
```

> **팁**
>
> 예외 메시지를 확인하면 문제를 파악하기 쉽습니다.
