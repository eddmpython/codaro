- 오늘 새로 배우는 개념: 복습

범위를 좁히는 이유는 간단합니다. 처음 배우는 사람은 한 번에 많은 문법을 보면 어디서 막혔는지 찾기 어렵습니다.

## 오늘의 학습 전략

- 연습 초점: 문자열, 리스트, 딕셔너리, 조건문, 반복문, 함수를 섞어 작은 프로그램을 만듭니다.
- 막히면: 어디서 막혔는지 문법 이름이 아니라 코드 줄 번호로 찾습니다.
