## 제너레이터 기초

*yield 키워드*

제너레이터는 yield 키워드를 사용하여 값을 하나씩 생성하는 함수입니다. return과 달리 yield는 함수 실행을 일시 중지하고 값을 반환한 후, 다음 호출 시 중지된 지점부터 계속 실행됩니다. 모든 값을 메모리에 저장하지 않고 필요할 때마다 생성하므로 매우 효율적입니다.

- yield: 값을 하나씩 생성
- 함수 실행 일시 중지
- 메모리 효율적
- 필요할 때 값 생성

### 기본 제너레이터

yield로 숫자를 하나씩 생성합니다.

```python
def simpleGen():
    yield 1
    yield 2
    yield 3

gen = simpleGen()
next(gen), next(gen), next(gen)
```

### 범위 제너레이터

특정 범위의 숫자를 생성합니다.

```python
def rangeGen(n):
    i = 0
    while i < n:
        yield i
        i = i + 1

g = rangeGen(5)
list(g)
```

### 제너레이터 순회

for 루프로 제너레이터를 순회합니다.

```python
def countdown(n):
    while n > 0:
        yield n
        n = n - 1

output = []
for val in countdown(5):
    output.append(val)
output
```

> **팁**
>
> 제너레이터는 한 번만 순회할 수 있습니다. 다시 순회하려면 새로 생성해야 합니다.

## 제너레이터 표현식

*간결한 제너레이터 생성*

제너레이터 표현식은 리스트 컴프리헨션과 비슷하지만 괄호를 사용합니다. (표현식 for 변수 in 시퀀스) 형태로 작성하며, 리스트를 만들지 않고 제너레이터를 반환합니다. 대량의 데이터를 처리할 때 메모리를 절약할 수 있습니다.

- (표현식 for 변수 in 시퀀스) 형식
- 괄호 사용
- 리스트 대신 제너레이터
- 메모리 절약

### 기본 제너레이터 표현식

제곱수를 생성하는 제너레이터를 만듭니다.

```python
squares = (x ** 2 for x in range(5))
list(squares)
```

### 조건부 제너레이터

조건을 만족하는 값만 생성합니다.

```python
evens = (n for n in range(10) if n % 2 == 0)
list(evens)
```

### 변환 제너레이터

문자열을 대문자로 변환합니다.

```python
words = ['hello', 'world', 'python']
upper = (w.upper() for w in words)
list(upper)
```

> **팁**
>
> 제너레이터 표현식은 sum(), max(), min() 같은 함수에 바로 전달할 수 있습니다.

## 이터레이터 기초

*iter()와 next()*

이터레이터는 순회 가능한 객체입니다. iter() 함수로 이터레이터를 생성하고, next() 함수로 다음 값을 가져옵니다. 리스트, 튜플, 문자열 등 모든 시퀀스는 iter()로 이터레이터를 만들 수 있으며, 더 이상 값이 없으면 StopIteration 예외가 발생합니다.

- iter(): 이터레이터 생성
- next(): 다음 값 가져오기
- StopIteration: 종료 신호
- 모든 시퀀스 지원

### 리스트 이터레이터

리스트를 이터레이터로 변환합니다.

```python
nums = [1, 2, 3]
it = iter(nums)
next(it), next(it), next(it)
```

### 문자열 이터레이터

문자열을 한 글자씩 순회합니다.

```python
text = 'abc'
textIter = iter(text)
next(textIter), next(textIter)
```

### 수동 순회

while과 next로 수동으로 순회합니다.

```python
items = [10, 20, 30]
iterator = iter(items)
collected = []
try:
    while True:
        collected.append(next(iterator))
except StopIteration:
    pass
collected
```

> **팁**
>
> for 루프는 내부적으로 iter()와 next()를 사용하여 순회합니다.

## 이터레이터 프로토콜

*__iter__와 __next__ 구현*

이터레이터 프로토콜은 __iter__와 __next__ 메서드를 구현하여 객체를 순회 가능하게 만드는 규약입니다. __iter__는 self를 반환하고, __next__는 다음 값을 반환하거나 StopIteration을 발생시킵니다. 이를 통해 사용자 정의 객체를 for 루프에서 사용할 수 있습니다.

- __iter__: self 반환
- __next__: 다음 값 반환
- StopIteration 발생
- for 루프 지원

### 기본 이터레이터 클래스

숫자를 순회하는 이터레이터를 만듭니다.

```python
class Counter:
    def __init__(self, max):
        self.max = max
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.max:
            self.current = self.current + 1
            return self.current
        raise StopIteration

c = Counter(3)
list(c)
```

### 범위 이터레이터

시작과 끝을 지정하는 이터레이터를 만듭니다.

```python
class Range:
    def __init__(self, start, end):
        self.current = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.end:
            val = self.current
            self.current = self.current + 1
            return val
        raise StopIteration

r = Range(5, 10)
list(r)
```

### 무한 이터레이터

무한히 값을 생성하는 이터레이터를 만듭니다.

```python
class Infinite:
    def __init__(self):
        self.num = 0

    def __iter__(self):
        return self

    def __next__(self):
        self.num = self.num + 1
        return self.num

inf = Infinite()
values = []
for val in inf:
    values.append(val)
    if val >= 5:
        break
values
```

> **팁**
>
> 제너레이터는 이터레이터 프로토콜을 자동으로 구현합니다.

## 고급 제너레이터

*실전 활용 패턴*

제너레이터는 파일 읽기, 무한 시퀀스 생성, 파이프라인 처리 등 다양한 실전 상황에서 활용됩니다. 메모리를 효율적으로 사용하면서도 코드를 간결하게 유지할 수 있어 대용량 데이터 처리에 매우 유용합니다.

- 파일 처리
- 무한 시퀀스
- 데이터 파이프라인
- 메모리 효율성

### 피보나치 제너레이터

피보나치 수열을 무한히 생성합니다.

```python
def fibonacci():
    a = 0
    b = 1
    while True:
        yield a
        a, b = b, a + b

fib = fibonacci()
sequence = []
for i in range(10):
    sequence.append(next(fib))
sequence
```

### 필터링 제너레이터

조건에 맞는 값만 생성합니다.

```python
def filterPositive(numbers):
    for n in numbers:
        if n > 0:
            yield n

dataset = [-2, -1, 0, 1, 2, 3]
filtered = filterPositive(dataset)
list(filtered)
```

### 변환 제너레이터

값을 변환하여 생성합니다.

```python
def doubleValues(nums):
    for n in nums:
        yield n * 2

source = [1, 2, 3, 4, 5]
doubled = doubleValues(source)
list(doubled)
```

> **팁**
>
> 여러 제너레이터를 연결하여 데이터 파이프라인을 만들 수 있습니다.

## 실전 패턴

*제너레이터와 이터레이터 활용*

제너레이터와 이터레이터는 실무에서 대용량 파일 처리, 배치 처리, 스트리밍 데이터 처리 등에 활용됩니다. 메모리를 절약하면서도 깔끔한 코드를 작성할 수 있어 파이썬의 강력한 기능 중 하나입니다.

- 배치 처리
- 청크 단위 읽기
- 파이프라인 구성
- 메모리 최적화

### 배치 제너레이터

데이터를 일정 크기로 묶어 생성합니다.

```python
def batchGen(elements, size):
    chunk = []
    for item in elements:
        chunk.append(item)
        if len(chunk) == size:
            yield chunk
            chunk = []
    if chunk:
        yield chunk

collection = [1, 2, 3, 4, 5, 6, 7]
batches = list(batchGen(collection, 3))
batches
```

### 체인 제너레이터

여러 제너레이터를 연결합니다.

```python
def square(values):
    for n in values:
        yield n ** 2

def addTen(integers):
    for n in integers:
        yield n + 10

original = [1, 2, 3, 4, 5]
pipeline = addTen(square(original))
list(pipeline)
```

### 조건부 생성

여러 조건을 조합하여 생성합니다.

```python
def complexFilter(entries):
    for item in entries:
        if item > 0 and item % 2 == 0:
            yield item * 3

integers = [-2, -1, 0, 1, 2, 3, 4, 5, 6]
outcomes = list(complexFilter(integers))
outcomes
```

> **팁**
>
> 제너레이터는 필요한 만큼만 값을 생성하므로 대용량 데이터 처리에 이상적입니다.
