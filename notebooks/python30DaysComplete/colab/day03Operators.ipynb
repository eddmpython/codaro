## 덧셈 연산자

*+ 기호로 더하기*

덧셈 연산자(+)는 두 숫자를 더합니다. 정수끼리, 소수끼리, 정수와 소수를 섞어서도 계산할 수 있습니다. 결과는 자동으로 화면에 표시됩니다.

- 정수 덧셈 가능
- 소수 덧셈 가능
- 정수 + 소수 = 소수

### 덧셈 계산

두 숫자를 더한 결과가 자동 출력됩니다.

```python
itemA = 10
itemB = 20
itemA + itemB
```

> **팁**
>
> 결과를 변수에 저장하려면 result = price + qty처럼 사용합니다.

## 뺄셈 연산자

*- 기호로 빼기*

뺄셈 연산자(-)는 앞 숫자에서 뒤 숫자를 뺍니다. 음수 결과도 가능하며, 순서가 중요합니다.

- 앞 숫자 - 뒤 숫자
- 음수 결과 가능
- 순서 바뀌면 결과도 바뀜

### 뺄셈 계산

100에서 35를 뺀 결과입니다.

```python
budget = 100
spent = 35
budget - spent
```

## 곱셈 연산자

** 기호로 곱하기*

곱셈 연산자(*)는 두 숫자를 곱합니다. 수학에서는 ×를 사용하지만, 프로그래밍에서는 *를 사용합니다.

- * 기호로 곱셈
- 정수와 소수 모두 가능
- 음수 곱셈도 가능

### 곱셈 계산

7과 8을 곱한 결과입니다.

```python
width = 7
height = 8
width * height
```

## 나눗셈 연산자

*/ 기호로 나누기*

나눗셈 연산자(/)는 앞 숫자를 뒤 숫자로 나눕니다. 결과는 항상 소수(float)로 나옵니다. 정수 나누기 정수도 결과는 소수입니다.

- / 기호로 나눗셈
- 결과는 항상 소수
- 0으로 나누면 에러

### 나눗셈 계산

10을 3으로 나눈 결과입니다.

```python
dividend = 10
divisor = 3
dividend / divisor
```

> **팁**
>
> 결과가 3.3333...처럼 소수로 나옵니다.

## 몫 연산자

*// 기호로 몫 구하기*

몫 연산자(//)는 나눗셈의 몫만 구합니다. 소수점 이하는 버리고 정수 부분만 반환합니다. 10 // 3 = 3 (나머지 1은 버림)

- // 기호로 몫 계산
- 소수점 이하 버림
- 정수 결과 반환

### 몫 계산

17을 5로 나눈 몫을 구합니다.

```python
items = 17
boxes = 5
items // boxes
```

## 나머지 연산자

*% 기호로 나머지 구하기*

나머지 연산자(%)는 나눗셈의 나머지를 구합니다. 10 % 3 = 1 (10을 3으로 나누면 몫 3, 나머지 1) 짝수/홀수 판별, 배수 확인 등에 자주 사용됩니다.

- % 기호로 나머지 계산
- 나눗셈 후 남은 값
- 짝수/홀수 판별에 유용

### 나머지 계산

17을 5로 나눈 나머지를 구합니다.

```python
coins = 17
slots = 5
coins % slots
```

> **팁**
>
> val % 2가 0이면 짝수, 1이면 홀수입니다.

## 거듭제곱 연산자

*** 기호로 제곱 계산*

거듭제곱 연산자(**)는 앞 숫자를 뒤 숫자만큼 거듭제곱합니다. 2 ** 3 = 8 (2 × 2 × 2) 수학의 2³를 프로그래밍에서는 2 ** 3으로 씁니다.

- ** 기호로 거듭제곱
- 2 ** 3 = 2³ = 8
- 제곱근도 가능 (9 ** 0.5 = 3)

### 거듭제곱 계산

2의 10승을 계산합니다.

```python
base = 2
power = 10
base ** power
```

## 같음 비교 연산자

*== 기호로 같은지 확인*

같음 연산자(==)는 두 값이 같은지 비교합니다. 같으면 True, 다르면 False를 반환합니다. 주의: 대입(=)과 비교(==)는 다릅니다!

- == 기호로 같은지 비교
- 결과는 True 또는 False
- = (대입)과 == (비교) 구별

### 같음 비교

두 숫자가 같은지 비교합니다.

```python
left = 10
right = 10
left == right
```

> **팁**
>
> = 하나는 대입, == 두 개는 비교입니다.

## 다름 비교 연산자

*!= 기호로 다른지 확인*

다름 연산자(!=)는 두 값이 다른지 비교합니다. 다르면 True, 같으면 False를 반환합니다.

- != 기호로 다른지 비교
- 다르면 True
- 같으면 False

### 다름 비교

두 숫자가 다른지 비교합니다.

```python
first = 5
second = 8
first != second
```

## 크기 비교 연산자

*> < >= <= 기호로 크기 비교*

크기 비교 연산자는 두 값의 대소를 비교합니다. > (크다), < (작다), >= (크거나 같다), <= (작거나 같다) 결과는 True 또는 False입니다.

- > (크다), < (작다)
- >= (크거나 같다), <= (작거나 같다)
- 결과는 True 또는 False

### 크다 (>)

왼쪽이 오른쪽보다 큰지 비교합니다.

```python
val = 15
target = 10
val > target
```

### 작다 (<)

왼쪽이 오른쪽보다 작은지 비교합니다.

```python
num = 15
limit = 10
num < limit
```

### 크거나 같다 (>=)

왼쪽이 오른쪽보다 크거나 같은지 비교합니다.

```python
level = 15
min = 10
level >= min
```

### 작거나 같다 (<=)

왼쪽이 오른쪽보다 작거나 같은지 비교합니다.

```python
stock = 15
max = 10
stock <= max
```

## and 논리 연산자

*모두 참이어야 참*

and 연산자는 양쪽 조건이 모두 True일 때만 True를 반환합니다. 하나라도 False면 결과는 False입니다. '그리고'의 의미입니다.

- 양쪽 모두 True → True
- 하나라도 False → False
- '그리고'의 의미

### and 연산

두 조건을 and로 결합합니다.

```python
adult = True
valid = True
adult and valid
```

## or 논리 연산자

*하나만 참이어도 참*

or 연산자는 양쪽 조건 중 하나만 True여도 True를 반환합니다. 둘 다 False일 때만 False입니다. '또는'의 의미입니다.

- 하나라도 True → True
- 둘 다 False → False
- '또는'의 의미

### or 연산

두 조건을 or로 결합합니다.

```python
cash = True
card = False
cash or card
```

## not 논리 연산자

*참과 거짓 반전*

not 연산자는 True를 False로, False를 True로 바꿉니다. 조건의 반대를 표현할 때 사용합니다. '아니다'의 의미입니다.

- True → False
- False → True
- '아니다'의 의미

### not 연산

조건을 반전시킵니다.

```python
rain = True
not rain
```

## in 멤버십 연산자

*포함 여부 확인*

in 연산자는 왼쪽 값이 오른쪽 문자열에 포함되어 있는지 확인합니다. 포함되어 있으면 True, 없으면 False를 반환합니다. 문자열에서 특정 문자나 단어를 찾을 때 사용합니다.

- 포함되어 있으면 True
- 없으면 False
- 문자열 검색에 유용

### in 연산

문자열에 특정 문자가 포함되어 있는지 확인합니다.

```python
lang = 'Python'
'th' in lang
```

> **팁**
>
> 'not in'을 사용하면 포함되지 않았는지 확인할 수 있습니다.

## 복합 대입 연산자

*연산과 대입을 한번에*

복합 대입 연산자는 연산과 대입을 동시에 합니다. x += 5는 x = x + 5와 같은 의미입니다. 코드를 더 간결하게 만들어줍니다.

- += 덧셈 후 대입
- -= 뺄셈 후 대입
- *= 곱셈 후 대입
- /= 나눗셈 후 대입

### 복합 대입 연산

숫자에 10을 더한 후 다시 저장합니다.

```python
balance = 50
balance += 10
balance
```

## 연산자 우선순위

*계산 순서 이해하기*

여러 연산자가 함께 사용될 때 계산 순서가 정해져 있습니다. 수학처럼 곱셈/나눗셈이 덧셈/뺄셈보다 먼저 계산됩니다. 괄호를 사용하면 우선순위를 변경할 수 있습니다.

- 1순위: 괄호 ()
- 2순위: 거듭제곱 **
- 3순위: 곱셈/나눗셈 * / // %
- 4순위: 덧셈/뺄셈 + -
- 5순위: 비교 연산자
- 6순위: 논리 연산자

### 연산자 우선순위

우선순위에 따라 계산됩니다.

```python
raw = 2 + 3 * 4
grouped = (2 + 3) * 4
print('without_paren:', raw)
print('with_paren:', grouped)
```

> **팁**
>
> 헷갈릴 때는 괄호를 사용하면 명확해집니다.
