## 오늘의 목표

인덱싱과 슬라이싱으로 문자열의 일부를 꺼냅니다.

## 왜 중요한가

데이터에서 필요한 부분만 꺼내는 능력은 파일명, 날짜, 코드, 문장 처리의 기초입니다.

## 생각 모델

문자열의 각 글자에는 0부터 시작하는 자리 번호가 있습니다. 슬라이싱은 시작 이상, 끝 미만의 범위를 잘라냅니다.

## 오늘 배우는 것

- 음수 인덱스로 뒤에서부터 접근하기
- 슬라이싱으로 부분 문자열 추출하기
- 스텝을 사용한 다양한 슬라이싱 패턴

## 학습 방법

1. 설명을 먼저 읽습니다.
2. 바로 아래 코드 셀을 실행합니다.
3. 출력이 설명과 어떻게 연결되는지 한 문장으로 말합니다.
4. 연습 셀에는 예제를 보지 않고 직접 다시 작성합니다.

## 오늘의 범위

- 오늘 새로 배우는 개념: 인덱싱, 음수 인덱싱, 슬라이싱, 간격 슬라이싱
- 이미 써도 되는 개념: 문자열 기초, 연산자
- 오늘은 일부러 쓰지 않는 개념: 문자열 메서드, 리스트, 직접 만드는 함수, import

범위를 좁히는 이유는 간단합니다. 처음 배우는 사람은 한 번에 많은 문법을 보면 어디서 막혔는지 찾기 어렵습니다.

## 오늘의 학습 전략

- 코딩 전: 글자 위에 0부터 번호를 적는다고 생각하고 꺼낼 위치를 먼저 표시합니다.
- 실행 중: 끝 번호의 글자는 포함되지 않는다는 점을 결과로 확인합니다.
- 연습 초점: 앞, 뒤, 중간, 간격을 바꾸며 같은 문자열을 여러 방식으로 잘라봅니다.
- 막히면: 결과가 한 글자 모자라면 끝 번호가 포함되지 않는 규칙을 다시 봅니다.
