- 인덱스는 0부터 시작
- 대괄호 []로 접근
- 첫 문자는 [0]

## 마지막 문자 접근

*길이를 이용한 마지막 문자*

마지막 문자의 인덱스는 문자열 길이 - 1입니다. len() 함수를 사용하여 길이를 구한 후 1을 빼면 마지막 인덱스를 얻을 수 있습니다.

- 마지막 인덱스 = 길이 - 1
- len() 함수 활용
- 범위를 벗어나면 에러

## 슬라이싱 기초

*[start:end]로 부분 문자열 추출*

슬라이싱은 문자열의 일부분을 추출합니다. [start:end] 형태로 사용하며, start 인덱스부터 end-1 인덱스까지 가져옵니다. end 위치는 포함되지 않는다는 점에 주의해야 합니다.

- [start:end] 형태
- start 포함, end 미포함
- 부분 문자열 추출

### 슬라이싱 기초

문자열의 일부를 추출합니다.

**실행 전**: 코드에서 값, 이름, 기호를 하나씩 짚고 어떤 결과가 나올지 먼저 예상하세요.

**실행 후**: 실제 결과가 어느 줄에서 만들어졌는지 한 문장으로 설명하세요.

```python
phrase = 'Python Programming'
phrase[0:6]
```

> **팁**
>
> [0:6]은 인덱스 0부터 5까지, 총 6개의 문자를 가져옵니다.

## 시작 생략 슬라이싱

*[:end]는 처음부터*

시작 인덱스를 생략하면 처음부터 추출합니다. [:end]는 [0:end]와 같은 의미입니다. 문자열의 앞부분을 가져올 때 편리합니다.

- [:end]는 처음부터 end-1까지
- [0:end]와 동일
- 앞부분 추출에 편리

### 시작 생략

**실행 후**: 실제 결과가 어느 줄에서 만들어졌는지 한 문장으로 설명하세요.

```python
msg = 'Hello World'
msg[:5]
```

- [start:]는 start부터 끝까지
- 마지막까지 추출
- 뒷부분 추출에 유용

### 끝 생략

## 전체 슬라이싱

*[:]는 전체 복사*

시작과 끝을 모두 생략하면 전체 문자열을 복사합니다. [:]는 원본과 같은 내용의 새로운 문자열을 만듭니다. 문자열 전체를 복사할 때 사용합니다.

- [:]는 전체 문자열
- 문자열 복사
- 원본과 동일한 내용

## 스텝 슬라이싱

- [start:end:step] 형태
- step은 간격
- 문자를 건너뛰며 추출

- [::-1]로 문자열 뒤집기
- 역순으로 추출
