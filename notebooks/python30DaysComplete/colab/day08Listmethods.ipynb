## 오늘의 목표

append(), insert(), remove(), pop(), sort() 등으로 리스트를 추가, 삭제, 정렬합니다.

## 왜 중요한가

리스트는 만들어 놓고 끝나는 값이 아니라 계속 변하는 작업 목록입니다.

## 생각 모델

리스트 메서드는 상자의 칸을 추가하거나 빼거나 순서를 바꾸는 동작입니다.

## 오늘 배우는 것

- append/insert로 요소 추가
- remove/pop으로 요소 삭제
- sort/reverse로 정렬과 뒤집기
- extend/copy로 리스트 확장과 복사

## 학습 방법

1. 설명을 먼저 읽습니다.
2. 바로 아래 코드 셀을 실행합니다.
3. 출력이 설명과 어떻게 연결되는지 한 문장으로 말합니다.
4. 연습 셀에는 예제를 보지 않고 직접 다시 작성합니다.

## 오늘의 범위

- 오늘 새로 배우는 개념: append(), insert(), remove(), pop(), sort(), reverse(), copy(), extend(), clear()
- 이미 써도 되는 개념: 리스트 기초
- 오늘은 일부러 쓰지 않는 개념: 튜플, 딕셔너리, 집합, 직접 만드는 함수, import

범위를 좁히는 이유는 간단합니다. 처음 배우는 사람은 한 번에 많은 문법을 보면 어디서 막혔는지 찾기 어렵습니다.

## 오늘의 학습 전략

- 코딩 전: 이 메서드가 리스트 자체를 바꾸는지, 새 값을 돌려주는지 먼저 예상합니다.
- 실행 중: 메서드 호출 전후 리스트를 출력해 무엇이 바뀌었는지 확인합니다.
- 연습 초점: 할 일 목록을 만들고 추가, 삽입, 제거, 정렬을 순서대로 적용합니다.
- 막히면: 원하는 값이 안 지워지면 값으로 지우는지 위치로 지우는지 구분합니다.
