## import 기본

*모듈 가져오기*

모듈은 함수와 변수를 담은 파이썬 파일입니다. import 모듈명으로 모듈을 가져오고, 모듈명.함수명()으로 사용합니다. 파이썬은 많은 표준 모듈을 제공합니다.

- import 모듈명 형식
- 모듈명.함수명() 사용
- 표준 라이브러리 제공
- 코드 재사용

### math 모듈

math 모듈을 가져와 사용합니다.

```python
import math

math.sqrt(16)
```

### 여러 함수 사용

모듈의 여러 함수를 사용할 수 있습니다.

```python
# import math  # 이미 위 셀에서 import한 경우 제거

radius = 5
area = math.pi * radius ** 2
area
```

### 여러 모듈

여러 모듈을 가져올 수 있습니다.

```python
# import math  # 이미 위 셀에서 import한 경우 제거
import random

value = math.ceil(3.2)
value
```

## from import

*특정 함수만 가져오기*

from 모듈명 import 함수명으로 특정 함수만 가져올 수 있습니다. 모듈명 없이 함수명만으로 사용할 수 있습니다. 여러 함수를 쉼표로 구분하여 가져올 수 있습니다.

- from 모듈명 import 함수명
- 모듈명 없이 사용
- 쉼표로 여러 함수
- 코드 간결화

### 함수 직접 사용

from import로 함수를 직접 사용합니다.

```python
from math import sqrt

sqrt(25)
```

### 여러 함수 가져오기

여러 함수를 한 번에 가져옵니다.

```python
from math import pi, ceil, floor

ceil(3.2), floor(3.8)
```

### 계산에 활용

가져온 함수와 상수를 계산에 사용합니다.

```python
from math import pi, pow

r = 3
volume = 4 / 3 * pi * pow(r, 3)
volume
```

> **팁**
>
> 필요한 함수만 가져오면 코드가 깔끔해집니다.

## as 별칭

*모듈/함수 이름 바꾸기*

as 키워드로 모듈이나 함수에 별칭을 붙일 수 있습니다. import 모듈명 as 별칭 또는 from 모듈명 import 함수명 as 별칭 형식입니다. 긴 이름을 짧게 만들거나 충돌을 피할 때 사용합니다.

- as 별칭 형식
- 긴 이름 단축
- 이름 충돌 방지
- 관습적 별칭 존재

### 모듈 별칭

모듈에 짧은 별칭을 붙입니다.

```python
import math as m

m.sqrt(36)
```

### 함수 별칭

함수에 별칭을 붙입니다.

```python
from math import sqrt as sq

sq(49)
```

### 여러 별칭

여러 함수에 별칭을 붙입니다.

```python
from random import randint as rInt, choice as pick

rInt(1, 10)
```

> **팁**
>
> 별칭은 짧고 의미있게 지으세요.

## math 모듈

*수학 함수*

math 모듈은 수학 함수를 제공합니다. sqrt(제곱근), ceil(올림), floor(내림), pow(거듭제곱), pi(파이), e(자연상수) 등을 사용할 수 있습니다.

- sqrt(): 제곱근
- ceil(): 올림
- floor(): 내림
- pow(): 거듭제곱

### 제곱근

sqrt로 제곱근을 계산합니다.

```python
from math import sqrt

sqrt(144)
```

### 올림과 내림

ceil과 floor로 올림과 내림을 합니다.

```python
from math import ceil, floor

num = 5.7
ceil(num), floor(num)
```

### 거듭제곱

pow로 거듭제곱을 계산합니다.

```python
from math import pow

pow(2, 10)
```

> **팁**
>
> math.pow는 실수를 반환하고, ** 연산자는 정수도 유지합니다.

## random 모듈

*난수 생성*

random 모듈은 난수를 생성합니다. randint(정수 난수), choice(리스트에서 선택), shuffle(리스트 섞기), random(0~1 실수) 등을 제공합니다.

- randint(a, b): a~b 정수
- choice(리스트): 랜덤 선택
- shuffle(리스트): 섞기
- random(): 0~1 실수

### 정수 난수

randint로 범위 내 정수를 생성합니다.

```python
from random import randint

randint(1, 100)
```

### 랜덤 선택

choice로 리스트에서 하나를 선택합니다.

```python
from random import choice

colors = ['red', 'green', 'blue', 'yellow']
choice(colors)
```

### 실수 난수

random()으로 0과 1 사이의 실수를 생성합니다.

```python
from random import random

random()
```

> **팁**
>
> 난수는 실행할 때마다 다른 값이 나옵니다.

## datetime 모듈

*날짜와 시간*

datetime 모듈은 날짜와 시간을 다룹니다. datetime.now(현재 시각), date(날짜), time(시간), timedelta(시간 차이) 등을 제공합니다.

- datetime.now(): 현재 시각
- date(): 날짜 생성
- time(): 시간 생성
- timedelta(): 시간 차이

### 현재 시각

datetime.now()로 현재 시각을 가져옵니다.

```python
from datetime import datetime

now = datetime.now()
now.year, now.month, now.day
```

### 날짜 생성

특정 날짜를 생성합니다.

```python
from datetime import date

birthday = date(2000, 1, 1)
birthday.year, birthday.month, birthday.day
```

### 시간 차이

timedelta로 시간 차이를 계산합니다.

```python
from datetime import date, timedelta

today = date(2024, 1, 1)
tomorrow = today + timedelta(days=1)
tomorrow.day
```

> **팁**
>
> datetime 객체는 여러 속성과 메서드를 제공합니다.
