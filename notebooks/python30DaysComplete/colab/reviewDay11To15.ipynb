## 리뷰 프로젝트

아래 코드는 해당 구간의 핵심 개념을 하나로 묶은 작은 프로젝트입니다. 먼저 실행하고, 그 다음 데이터와 변수명을 바꿔 다시 작성하세요.

```python
records = [
    {"topic": "dict", "score": 85},
    {"topic": "loop", "score": 72},
    {"topic": "function", "score": 91},
]
def grade(score):
    if score >= 90:
        return "A"
    if score >= 80:
        return "B"
    return "C"

report = {}
for record in records:
    report[record["topic"]] = grade(record["score"])
report
```
