- 오늘 새로 배우는 개념: print() 출력, 주석, 문자열 값

## 오늘의 학습 전략

- 연습 초점: 문장을 바꾸고 줄 수를 바꾸어도 실행 결과를 말로 설명합니다.
- 막히면: 오류가 나면 따옴표, 괄호, 줄바꿈을 먼저 확인합니다.
