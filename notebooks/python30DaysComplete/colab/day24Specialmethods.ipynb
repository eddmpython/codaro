특수 메서드로 객체가 출력, 길이, 더하기, 비교, 인덱싱에 반응하게 만듭니다.

## 왜 중요한가

객체가 파이썬 기본 문법과 자연스럽게 어울리면 사용하기 쉬운 코드를 만들 수 있습니다.

## 생각 모델

특수 메서드는 파이썬이 특정 기호나 내장 함수 뒤에서 몰래 호출하는 약속된 이름입니다.

## 오늘 배우는 것

- __str__과 __repr__로 문자열 표현
- __len__으로 길이 정의
- __add__, __eq__로 연산자 구현
- __getitem__으로 인덱싱 지원

## 학습 방법

2. 바로 아래 코드 셀을 실행합니다.

4. 연습 셀에는 예제를 보지 않고 직접 다시 작성합니다.

## 오늘의 범위

- 오늘 새로 배우는 개념: __str__, __repr__, __len__, __add__, __eq__, __getitem__
- 이미 써도 되는 개념: 클래스 전체
- 오늘은 일부러 쓰지 않는 개념: 데코레이터, 메타클래스

범위를 좁히는 이유는 간단합니다. 처음 배우는 사람은 한 번에 많은 문법을 보면 어디서 막혔는지 찾기 어렵습니다.

## 오늘의 학습 전략

- 코딩 전: 어떤 문법이 어떤 특수 메서드로 연결되는지 먼저 짝을 맞춥니다.
- 실행 중: print(obj), len(obj), obj + other 같은 호출이 어떤 메서드를 실행하는지 확인합니다.
- 연습 초점: 내가 만든 객체가 문자열처럼 출력되고 리스트처럼 일부 값을 꺼내게 만듭니다.
- 막히면: 결과가 이상하면 특수 메서드가 반드시 값을 return하는지 확인합니다.
